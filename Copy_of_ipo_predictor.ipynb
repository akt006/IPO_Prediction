```python
missing_columns = ["HNI", "RII", "TOTAL", "QIB"]
```

```python
def random_imputation(data, feature):

    number_missing = data[feature].isnull().sum()
    observed_values = data.loc[data[feature].notnull(), feature]
    data.loc[data[feature].isnull(), feature + '_imp'] = np.random.choice(observed_values, number_missing, replace = True)
    return data
```

```python
# here we are new coloumnn is made and imputeed data is entered in that the code effectively performs random imputation for all missing values in each feature and stores the imputed values in a new feature column with the same name plus '_imp'. This technique of creating a new feature column to store imputed values is often used to preserve the original data and allow for comparison between the original and imputed datasets.
for feature in missing_columns:
    data[feature + '_imp'] = data[feature]
    data = random_imputation(data, feature)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

# knn
knn = KNeighborsRegressor(n_neighbors=10)

dmae = []
rmse = []

# raw, normalized and standardized training and testing data
trainX = [train_x, train_x_norm_df]
testX = [val_x , val_x_norm_df]

# model fitting and measuring RMSE
for i in range(len(trainX)):

    # fit
    knn.fit(trainX[i],train_y)
    acc1 =knn.score(testX[i], val_y)
    print (acc1)
    # predict
    pred = knn.predict(testX[i])
    # MAE
    dmae.append(mean_absolute_error(val_y,pred))
    rmse.append(np.sqrt(mean_squared_error(val_y,pred)))

# visualizing the result
df_knn = pd.DataFrame({'MAE':dmae},index=['Original','Normalized'])
df_1knn = pd.DataFrame({'RMSE':rmse},index=['Original','Normalized'])
print(df_knn)
print(df_1knn)
```
